# Machine Learning Models Implementation

## Logistic Regression Implementation
Logistic Regression is essentially a linear model that uses the sigmoid function to output probabilities.

 implementing a logistic regression model using gradient descent for training, where weights and bias are optimized to minimize the logistic loss. The sigmoid function is used to transform linear model outputs into probabilities, and the predict method converts these probabilities into binary predictions.

```python
import numpy as np

class LogisticRegressionCustom:
    def __init__(self, learning_rate=0.01, num_iterations=1000, threshold=0.5):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.threshold = threshold

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        # Number of samples and features
        m, n = X.shape

        # Initialize weights and bias
        self.weights = np.zeros(n)
        self.bias = 0

        # Gradient descent
        for _ in range(self.num_iterations):
            model = np.dot(X, self.weights) + self.bias
            predictions = self.sigmoid(model)

            # Compute the gradient
            dw = (1/m) * np.dot(X.T, (predictions - y))
            db = (1/m) * np.sum(predictions - y)

            # Update weights and bias
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        model = np.dot(X, self.weights) + self.bias
        predictions = self.sigmoid(model)
        return np.where(predictions >= self.threshold, 1, 0)

    def get_params(self, deep=True):
        return {
            'learning_rate': self.learning_rate,
            'num_iterations': self.num_iterations,
            'threshold': self.threshold
        }

    def set_params(self, **params):
        for param, value in params.items():
            setattr(self, param, value)
        return self



```

## Evaluation metrics

```python
from sklearn.model_selection import train_test_split

def accuracy_score(y_true, y_pred):
    return np.mean(y_true == y_pred)

def precision_score(y_true, y_pred):
    true_positive = np.sum((y_pred == 1) & (y_true == 1))
    predicted_positive = np.sum(y_pred == 1)
    return true_positive / predicted_positive if predicted_positive != 0 else 0

def recall_score(y_true, y_pred):
    true_positive = np.sum((y_pred == 1) & (y_true == 1))
    actual_positive = np.sum(y_true == 1)
    return true_positive / actual_positive if actual_positive != 0 else 0

def f1_score(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    return 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

def confusion_matrix(y_true, y_pred):
    tp = np.sum((y_true == 1) & (y_pred == 1))
    tn = np.sum((y_true == 0) & (y_pred == 0))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))
    return np.array([[tn, fp], [fn, tp]])

def evaluate_model(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    cm = confusion_matrix(y_true, y_pred)

    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1-Score:", f1)
    print("Confusion Matrix:\n", cm)
```

## Most important evaluation metric:

**Recall (or Sensitivity)** would be the most critical evaluation metric in this scenario for the following reasons:

Maximizing Reach: we want to identify as many potential customers interested in caravan insurance as possible. High recall ensures that you are capturing most of the potential interested customers, even if it means some of the identified customers might not be interested.

Cost of False Negatives: If failing to identify a genuinely interested customer (false negative) has a higher cost than incorrectly targeting a non-interested customer (false positive), then recall becomes crucial.

```python
from sklearn.model_selection import train_test_split


# Split the dataset into features and target
X = train_data.iloc[:, :-1].values
y = train_data['CARAVAN'].values

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Logistic Regression
logistic_model = LogisticRegressionCustom(learning_rate=0.01, num_iterations=10000,threshold=0.5)
logistic_model.fit(X_train, y_train)
logistic_predictions = logistic_model.predict(X_test)
print("Logistic Regression Evaluation:")
evaluate_model(y_test, logistic_predictions)
```

## Class Imbalance
The results we're seeing indicate that while the models achieve high accuracy, they fail to correctly predict any of the positive class (1). This often occurs when the dataset is highly imbalanced.

## Threshold Issue:

## Hyperparameter tuning to find the best parameters for the model using grid search:
method for hyperparameter tuning that involves systematically evaluating all possible combinations of specified hyperparameters to determine the optimal settings for a machine learning model.

```python
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Define the parameter grid
param_grid = {
    'learning_rate': [0.001, 0.01, 0.1],
    'num_iterations': [1000, 5000, 10000],
    'threshold': [0.3, 0.5, 0.7]
}

# Create the GridSearchCV object
grid_search = GridSearchCV(
    estimator=LogisticRegressionCustom(),
    param_grid=param_grid,
    scoring='f1',  # To optimize for accuracy and recall
    cv=5,  # Number of cross-validation folds
    verbose=1,
    n_jobs=-1  # Use all available cores
)

# Split the dataset into features and target
X = train_data.iloc[:, :-1].values
y = train_data['CARAVAN'].values

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Perform grid search
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation score: ", grid_search.best_score_)

# Evaluate the best model on the test set
best_model = grid_search.best_estimator_
logistic_predictions = best_model.predict(X_test)
evaluate_model(y_test, logistic_predictions)
```

## Evaluation:
- High Accuracy with Low Recall: The model is achieving high accuracy largely due to the high number of true negatives and a low number of true positives and false negatives. This often occurs in imbalanced datasets where the majority class dominates.
- Low Precision and Recall: The low recall suggests that the model struggles to identify positive cases, which is critical for this problem.
- The low cross-validation score suggests that the models performance is not consistent across different subsets of the data or might be struggling with certain configurations.

## Next steps:

using ensemble model like random forest can often yield better results, especially in cases where a single model is struggling due to issues like class imbalance

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Train a Random Forest model
def train_random_forest(X_train, y_train, n_estimators=100, max_depth=None, random_state=42):
    rf_model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=random_state)
    rf_model.fit(X_train, y_train)
    return rf_model

# Evaluate the model using test data
def evaluate_model_sklearn(model, X_test, y_test):
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1-Score:", f1)
    print("Confusion Matrix:\n", cm)


# Train Random Forest on the resampled training data
rf_model = train_random_forest(X, y, n_estimators=100, max_depth=10)

# Evaluate the Random Forest model on the test data
print("Random Forest Evaluation:")
evaluate_model_sklearn(rf_model, X_test, y_test)
```

Accuracy (93.99%): The accuracy is high, indicating that the model is correctly predicting the majority of the cases. However, accuracy alone can be misleading in imbalanced datasets.

Precision (1.0): This means that when the model predicts the positive class (1), it is correct 100% of the time. However, this is likely due to the model making very few positive predictions.

Recall (11.76%): The recall is quite low, indicating that the model only identifies about 11.76% of the actual positive cases. This suggests that the model is heavily biased towards predicting the negative class (0).

F1-Score (0.21): The F1-score is low, indicating that the balance between precision and recall is not optimal. This is expected given the low recall.

Confusion Matrix:

True Negatives (TN): 1628 (correctly predicted 0)
False Positives (FP): 0 (no incorrect predictions of 1)
False Negatives (FN): 105 (missed 1 predictions)
True Positives (TP): 14 (correctly predicted 1)

Interpretation:
High Precision but Low Recall: The Random Forest model is very cautious in predicting the positive class, resulting in high precision but low recall. This often happens when the decision threshold for classifying a positive case is set high, causing the model to err on the side of predicting the negative class.

## Hyperparamter tuning for the random forest model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Define the parameter distribution
param_dist = {
    'n_estimators': randint(100, 500),      # Number of trees, using a range
    'max_depth': [None, 10, 20, 30],        # Maximum depth of each tree
    'min_samples_split': randint(2, 11),    # Minimum number of samples to split an internal node
    'min_samples_leaf': randint(1, 5),      # Minimum number of samples at a leaf node
    'class_weight': [{0: 1, 1: 10}, {0: 1, 1: 5}, {0: 1, 1: 20}]  # Different class weights
}

# Initialize the Random Forest model
rf_model = RandomForestClassifier(random_state=42)

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=rf_model, param_distributions=param_dist,
                                   n_iter=100, cv=5, scoring='f1', verbose=2, random_state=42, n_jobs=-1)

# Fit the random search to the data (using the resampled training data)
random_search.fit(X, y)

# Extract the best model and parameters
best_rf_model = random_search.best_estimator_
best_params = random_search.best_params_

print("Best parameters found by RandomizedSearchCV:")
print(best_params)

# Evaluate the best model on the test set
print("Evaluation of the best Random Forest model:")
evaluate_model_sklearn(best_rf_model, X_test, y_test)
```

If the cost of reaching out to uninterested customers is low, the high recall justifies the model's use because it ensures you capture nearly all potential customers.
However, if false positives (uninterested customers) are costly in terms of resources, we may want to improve precision, possibly by adjusting the class weights further or exploring other models or techniques like boosting.

Boosting algorithms, such as Gradient Boosting often excel in handling imbalanced datasets. By focusing on the misclassified samples in each iteration, boosting methods can gradually shift their attention to the minority class (customers interested in caravan insurance in our case), improving their detection.

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV, train_test_split
import numpy as np

X = train_data.drop(columns=['CARAVAN'])  # Features (dropping the target column)
y = train_data['CARAVAN']  # Target variable

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the parameter grid to search
param_grid = {
    'n_estimators': np.arange(50, 300, 50),  # Number of boosting stages
    'learning_rate': np.linspace(0.01, 0.3, 10),  # Learning rate shrinks the contribution of each tree
    'max_depth': np.arange(3, 15, 2),  # Maximum depth of the individual trees
    'min_samples_split': np.arange(2, 20, 2),  # Minimum number of samples required to split an internal node
    'min_samples_leaf': np.arange(1, 20, 2),  # Minimum number of samples required to be at a leaf node
    'subsample': np.linspace(0.6, 1.0, 5),  # Fraction of samples used for fitting individual base learners
    'max_features': ['auto', 'sqrt', 'log2', None],  # Number of features to consider when looking for the best split
}

# Initialize the Gradient Boosting model
gb_model = GradientBoostingClassifier(random_state=42)

# Set up the RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=gb_model,
                                   param_distributions=param_grid,
                                   n_iter=50,  # Number of parameter settings that are sampled
                                   scoring='f1',  # Scoring metric to evaluate the model
                                   cv=5,  # Number of cross-validation folds
                                   verbose=2,
                                   random_state=42,
                                   n_jobs=-1)  # Use all available cores

# Fit the RandomizedSearchCV to find the best parameters
random_search.fit(X, y)

# Print the best parameters found
print("Best parameters found: ", random_search.best_params_)

# Train the Gradient Boosting model with the best parameters on the entire training set
best_gb_model = random_search.best_estimator_
best_gb_model.fit(X, y)

# Predict on the test set
y_pred = best_gb_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Print evaluation results
print("Gradient Boosting Evaluation (with Hyperparameter Tuning):")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
print("Confusion Matrix:\n", conf_matrix)
```

The Gradient Boosting model performs well, with high accuracy (98.51%) and precision (96.04%), meaning it effectively identifies interested customers with few false positives. However, the recall is slightly lower at 81.51%, indicating some potential customers are missed. The model strikes a good balance between precision and recall, as reflected in the F1-Score of 0.8818.
